```python
from pydataset import data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score as acs
```

```python
pima = data("Pima.tr")
pima.plot(kind="scatter", x='skin', y='bmi', edgecolor="#1D2B53")
```

```python
X_train, X_test, y_train, y_test = train_test_split(pima.skin, pima.bmi)
plt.scatter(X_train, y_train, label="training data", color='r', alpha=0.7, edgecolors="k")
plt.scatter(X_test, y_test, label="testing data", color='b', alpha=0.7, edgecolors="k")
plt.xlabel("skin")
plt.ylabel("bmi")
plt.legend()
plt.show()
```

```python
model = LinearRegression()
model.fit(X_train.values.reshape(-1,1), y_train.values)

prediction = model.predict(X_test.values.reshape(-1,1))
plt.plot(X_test, prediction, label="LinearRegression", color='#77fc03')
plt.scatter(X_test, y_test, label="actual test", color="b", alpha=0.7)
plt.legend()
plt.show()
```

```python
model.predict(np.array([[40]]))[0]
```

```python
model.score(X_test.values.reshape(-1,1), y_test.values)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

#  predictions
prediction = model.predict(X_test.values.reshape(-1,1))

# Mean Absolute Error
mae = mean_absolute_error(y_test, prediction)

# Mean Squared Error
mse = mean_squared_error(y_test, prediction)

#  R-squared
r2 = r2_score(y_test, prediction)

print("Mean Absolute Error:", mae)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
```
